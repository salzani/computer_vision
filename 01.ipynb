# Starting with OCR

```python
import pytesseract
import numpy as np
import cv2 # OpenCV
import matplotlib.pyplot as plt
```

**Importing an image**

```python
first_img = cv2.imread('text-recognize/Imagens/Aula1-teste.png')
second_img = cv2.imread('text-recognize/Imagens/Aula1-ocr.png')
```

## Showing the images

```python
first_img_rgb = cv2.cvtColor(first_img, cv2.COLOR_BGR2RGB)
second_img_rgb = cv2.cvtColor(second_img, cv2.COLOR_BGR2RGB)
second_img_gray = cv2.cvtColor(second_img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('First Image')
plt.imshow(first_img_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Second Image')
plt.imshow(second_img_rgb)
plt.axis('off')

plt.show()
```

**Extracting a text from the image**

```python
text_01 = pytesseract.image_to_string(first_img_rgb)

print(f'Image text -> {text_01}')
```

```python
text_01_ocr = pytesseract.image_to_string(second_img_rgb)
print(f'Image text: \n{text_01_ocr}')

text_01_ocr_gray = pytesseract.image_to_string(second_img_gray)
print(f'Image text: \n{text_01_ocr_gray}')
```
