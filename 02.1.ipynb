````python
import pytesseract
from pytesseract import Output
import numpy as np
import cv2 # OpenCV
import matplotlib.pyplot as plt
import os
````

````python
first_img = cv2.imread('text-recognize/Imagens/Aula2-undersampling.png')

first_img_rgb = cv2.cvtColor(first_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('First Image')
plt.imshow(first_img_rgb)
plt.axis('off')

plt.show()
````

````python
text_01 = pytesseract.image_to_string(first_img_rgb)

print(f'Image text: \n{text_01}')
````

# *The correct PSM for the type of image you are analyzing is essential for accurate analysis*



````python
config_tesseract = 'tessdata-dir tessdata --psm 6'

txt = pytesseract.image_to_string(rgb, config=config_tesseract, lang='por')

print(txt)
````
